```python

import pandas as pd
import cherrypy as cpy 
import json
import os


class MTA_serviceApp_Web():
    
    def __init__(self):
        
        self.usr_dct = {}
        self.usr_srt = ''
        # self.users_df = pd.DataFrame()
        
        if os.path.isfile("UsrData/UsrDataFile.txt"):
            self.f1 = open("UsrData/UsrDataFile.txt", 'r+')
            try: 
                self.js1 = json.load(self.f1)
                self.usr_dct.update(self.js1)
            except:
                print('no file returned')
                pass
            
            # self.js1 = json.load(self.f1)
            
            
        else:
            self.f1 = open('UsrData/UsrDataFile.txt', 'w+')
        
        
        
        
    def index(self):
                
        
        html_fang = """
        <html>
          <head>
              <title>Sign-up Page</title>
              <link href="/static/css/style.css" rel="stylesheet">      
          </head>
          <body><fieldset><legend>MTA Notification</legend>
          <form method="get" action="updt_usr_string" target="_self"><h1>Sign up</h1><br/><p>Enter your email address here: <br>
             <input type="email" name="usr_email" placeholder="Email address" id="email" required></p><p> When do you take MTA? Enter time in 24hr format: <br>
             <input type="text"  name="usr_commute_time" placeholder="HH:MM" pattern="([01]?[0-9]|2[0-3]):[0-5][0-9]" id="time" required></p><p><fieldset><legend>Please select all the possible lines you take: </legend>
                 <input type="checkbox" name="use_123" value="123"/><lable for="123">1,2,3</lable><br />
                 <input type="checkbox" name="use_456" value="456"/><label for="456">4,5,6</label><br />
                 <input type="checkbox" name="use_7" value="7"/><label for="7">7</label><br />
                 <input type="checkbox" name="use_ACE" value="ACE"/><label for="ACE">A,C,E</label><br />
                 <input type="checkbox" name="use_BDFM" value="BDFM"/><label for="BDFM">B,D,F,M</label><br />
                 <input type="checkbox" name="use_G" value="G"/><label for="G">G</label><br />            
                 <input type="checkbox" name="use_JZ" value="JZ"/><label for="JZ">J,Z</label><br />  
                 <input type="checkbox" name="use_L" value="L"/><label for="L">L</label><br />
                 <input type="checkbox" name="use_NQR" value="NQR"/><label for="NQR">N,Q,R</label><br />
                 <input type="checkbox" name="use_S" value="S"/><label for="S">S</label><br />
                 <input type="checkbox" name="use_SIR" value="SIR"/><label for="SIR">S,I,R</label><br /></fieldset><p>
             <button type="submit">submit your info</button></p>
            </form></fieldset>
           </body>
           <?php
           ?>
         </html>"""
        
        return html_fang
    
    def updt_usr_string(self,usr_email,usr_commute_time,use_123=0,use_456=0,use_7=0,use_ACE=0,use_BDFM=0,use_G=0,use_JZ=0,use_L=0,use_NQR=0,use_S=0,use_SIR=0):
        self.usr_dct.update({ usr_email : [ usr_commute_time, use_123,use_456,use_7,use_ACE,use_BDFM,use_G,use_JZ,use_L,use_NQR,use_S,use_SIR] } )
        
        
        html = """
            <html><body>
            You just inserted commuter info for %s.<br /><br />
            
            <a href="/">Click here to go back to main page</a><br />
            <a href="/shutdown">Click here if you are done entering user info.</a>
            
            </body></html>
            """ % (usr_email)
        
        return html
        
        
        
    def shutdown(self):
        
        self.f1 = open("UsrData/UsrDataFile.txt", 'r+')
        self.f1.truncate()
        json.dump(self.usr_dct,self.f1)
        self.f1.close()
        
        cpy.engine.exit()
        
        return """
        <html><body>
            Thank you for using the MTA Service Status Service
            
        </body></html>
        """
    
    def curr_user_dict(self):
        curr_dict = self.usr_dct
        usr_data_json = json.dumps(curr_dict)
        return usr_data_json
    
    
    index.exposed = True
    shutdown.exposed = True
    curr_user_dict.exposed = True
    updt_usr_string.exposed = True
    
        

mta_appRun = MTA_serviceApp_Web()
cpy.quickstart(mta_appRun)

        
```
